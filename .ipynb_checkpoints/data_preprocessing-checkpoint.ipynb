```python
from scipy.spatial import ConvexHull
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['axes.labelsize'] = 17
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
%matplotlib inline 
from skimage.io import imread
from skimage import color
import copy
import cv2
from skimage.feature import canny
from skimage.transform import hough_line, hough_line_peaks
import skimage.filters
```

```python
def resize_image_and_label(image, label, length = 64):
    # shift label
    if image.shape[2] == 4:  # Image has 4 channels (RGBA)
        # Convert to RGB by ignoring the alpha channel
        image = image[:, :, :3]
    # make the image a square
    Height = np.shape(image)[0]
    Width = np.shape(image)[1]
    Area = Height*Width
    diff = abs(Width - Height)
    # Determine padding for width and height
    if Width > Height:
        # Width is greater than height, pad top and bottom
        top_pad = diff// 2
        bottom_pad = diff - top_pad
        left_pad = 0
        right_pad = 0
    else:
        # Height is greater than width, pad left and right
        left_pad = diff // 2
        top_pad = 0
        bottom_pad = 0
        right_pad = diff - left_pad
    # Apply padding
    image = cv2.copyMakeBorder(image, top_pad, bottom_pad, left_pad, right_pad, cv2.BORDER_CONSTANT, value=[255, 255, 255])
    
    if diff > 0:
        label = np.copy(label)
        for i in range(len(label)):
            # Adjusting x coordinates for left padding
            label[i][0] += left_pad
            # Adjusting y coordinates for top padding
            label[i][1] += top_pad
    # rescaled dimensions
    Height = np.shape(image)[0]
    Width = np.shape(image)[1]
    Area = Height*Width
    area = length*length
    width  = int(np.sqrt(area/Area) * Width)
    height = int(np.sqrt(area/Area) * Height)
    image = cv2.resize(image, (length,length))
    label = label * np.sqrt(area/Area)
    for l in label:
        for i in l:
            i = int(i)
    
    return image, label

def adjust_tracked_label(label,i):
    #1080x1920
    tracked_index = [i for i in range(489,643)]
    for j in range(757,936):
        tracked_index.append(j)
    if i in tracked_index:
        l = copy.copy(label)
        label[:, 0] = l[:, 1] #+ int(1080/2)
        label[:, 1] = l[:, 0] #- int(1920/2)
    return label
```

```python
def face_center(label, face=1):
    if face == 1:
        points = [0, 1, 2, 3]
        line_1 = [0, 2]
        line_2 = [1, 3]
    elif face == 2:
        points = [0, 3, 4, 5]
        line_1 = [0, 4]
        line_2 = [3, 5]
    elif face == 3:
        points = [0, 5, 6, 1]
        line_1 = [0, 6]
        line_2 = [1, 5]
    
    # Extract points
    p = np.array([label[i,:] for i in points])
    q = np.array([label[i,:] for i in range(7)])
    # Calculate center point
    c = np.mean(p, axis=0)
    
    # Calculate intersection point
    intersection = find_intersection(q[line_1[0]], q[line_1[1]], q[line_2[0]], q[line_2[1]])
    
    # Check if c coincides with intersection point
    if np.linalg.norm(intersection - c)< np.linalg.norm(q[line_1[1]]-q[line_1[0]])*0.05:
        return True
    else:
        return False
    
def find_intersection(p1, p2, p3, p4):
    x1, y1 = p1[0], p1[1]
    x2, y2 = p2[0], p2[1]
    x3, y3 = p3[0], p3[1]
    x4, y4 = p4[0], p4[1]
    
    # Calculate slopes
    m1 = (y2 - y1) / (x2 - x1) if x2 != x1 else np.inf
    m2 = (y4 - y3) / (x4 - x3) if x4 != x3 else np.inf
    
    # Check if lines are parallel
    if np.isinf(m1) and np.isinf(m2):
        return np.nan, np.nan
    elif np.isinf(m1):
        x = x1
        y = m2 * (x - x3) + y3
    elif np.isinf(m2):
        x = x3
        y = m1 * (x - x1) + y1
    else:
        # Calculate intersection point
        x = (m1*x1 - m2*x3 + y3 - y1) / (m1 - m2)
        y = m1 * (x - x1) + y1
    
    return np.array([x, y])

def sort_label(label):
    # determine inner corner of the cube
    # this point is inside the convex hull
    hull = ConvexHull(label)
    hull_vertices = hull.vertices
    point_not_on_hull = None
    for i, point in enumerate(label):
        if i not in hull_vertices:
            point_not_on_hull = point
            break
    center = point_not_on_hull
    # Sort points clockwise relative to the center
    sorted_label = sorted(label, key=lambda p: -(np.arctan2(p[1] - center[1], p[0] - center[0]) - 2 * np.pi) % (2 * np.pi))
    if not face_center(np.array(sorted_label), face=1):
        s = [sorted_label[0]]
        for i in range(6):
            s.append(sorted_label[1:][(i-1)])
        sorted_label = s
    # take (x,y) coordinates and split them
    result = []
    for y in sorted_label:
        result.append(y[0])
        result.append(y[1])
    return np.array(result)

def reconstruct_label(label):
    import numpy as np
    return np.array([(label[i], label[i+1]) for i in range(0, len(label), 2)])
```

```python
def import_data(i_max = 972, i_min = 1, sort_label = sort_label):
    formats = [".jpg", "png",".jpeg","",".png","JPG"]
    images = []
    labels = []

    for i in range(i_min,i_max):
        for fmt in formats:
            try:
                image = imread('dataset/pics/'+str(i)+fmt)
            
            except:
                continue
        label = np.loadtxt('dataset/labels/'+str(i)+".txt")
        label = adjust_tracked_label(label,i)
        image, label = resize_image_and_label(image, label, length = 224)
        label = sort_label(label)
        images.append(image)
        labels.append(label)
    images = np.array(images)
    return images, labels
```

```python
def flip_horizontally(image, label):
    height, width = image.shape[:2]
    flipped_image = np.fliplr(image)
    flipped_label = np.array([(width - x, y) for (x, y) in label])
    return flipped_image, flipped_label

def flip_vertically(image, label):
    height, width = image.shape[:2]
    flipped_image = np.flipud(image)
    flipped_label = np.array([(x, height - y) for (x, y) in label])
    return flipped_image, flipped_label

def flip_diagonally(image, label):
    flipped_image = np.transpose(image, axes=(1, 0, 2))  # Transpose only first two dimensions
    flipped_label = np.array([(y, x) for (x, y) in label])
    return flipped_image, flipped_label

def flip_anti_diagonally(image, label):
    flipped_image = np.flipud(image)
    flipped_image = np.transpose(flipped_image, axes=(1, 0, 2))  # Transpose only first two dimensions
    flipped_label = np.array([(image.shape[0] - y, x) for (x, y) in label])
    return flipped_image, flipped_label
```

```python
images, labels = import_data(i_max = 498, i_min = 1)
```

```python
i = np.random.randint(0,498)
label = labels[i]
image = images[i]
flipped_horizontal_image, flipped_horizontal_label = flip_horizontally(image, reconstruct_label(label))
flipped_vertical_image, flipped_vertical_label = flip_vertically(image, reconstruct_label(label))
flipped_diagonal_image, flipped_diagonal_label = flip_diagonally(image, reconstruct_label(label))
flipped_anti_diagonal_image, flipped_anti_diagonal_label = flip_anti_diagonally(image, reconstruct_label(label))

flipped_horizontal_label = reconstruct_label(sort_label(flipped_horizontal_label))

flipped_vertical_label = reconstruct_label(sort_label(flipped_vertical_label))

flipped_diagonal_label = reconstruct_label(sort_label(flipped_diagonal_label))

flipped_anti_diagonal_label = reconstruct_label(sort_label(flipped_anti_diagonal_label))
```

```python

# Plotting
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Original image and label
axes[0, 0].imshow(image, cmap='gray')
for i in range(7):
    axes[0, 0].scatter(reconstruct_label(label)[i, 0], reconstruct_label(label)[i, 1], color='blue')
    axes[0, 0].text(reconstruct_label(label)[i, 0], reconstruct_label(label)[i, 1], str(i), fontsize=12, color='blue')
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')
# Flipped horizontally
axes[0, 1].imshow(flipped_horizontal_image, cmap='gray')
for i in range(7):
    axes[0, 1].scatter(flipped_horizontal_label[i, 0], flipped_horizontal_label[i, 1], color='blue')
    axes[0, 1].text(flipped_horizontal_label[i, 0], flipped_horizontal_label[i, 1], str(i), fontsize=12, color='blue')

axes[0, 1].set_title('Flipped Horizontally')
axes[0, 1].axis('off')
# Flipped vertically
axes[0, 2].imshow(flipped_vertical_image, cmap='gray')
for i in range(7):
    axes[0, 2].scatter(flipped_vertical_label[i, 0], flipped_vertical_label[i, 1], color='blue')
    axes[0, 2].text(flipped_vertical_label[i, 0], flipped_vertical_label[i, 1], str(i), fontsize=12, color='blue')
axes[0, 2].set_title('Flipped Vertically')
axes[0, 2].axis('off')
# Flipped diagonally
axes[1, 0].imshow(flipped_diagonal_image, cmap='gray')
for i in range(7):
    axes[1, 0].scatter(flipped_diagonal_label[i, 0], flipped_diagonal_label[i, 1], color='blue')
    axes[1, 0].text(flipped_diagonal_label[i, 0], flipped_diagonal_label[i, 1], str(i), fontsize=12, color='blue')
axes[1, 0].set_title('Flipped Diagonally')
axes[1, 0].axis('off')
# Flipped anti-diagonally
axes[1, 1].imshow(flipped_anti_diagonal_image, cmap='gray')
for i in range(7):
    axes[1, 1].scatter(flipped_anti_diagonal_label[i, 0], flipped_anti_diagonal_label[i, 1], color='blue')
    axes[1, 1].text(flipped_anti_diagonal_label[i, 0], flipped_anti_diagonal_label[i, 1], str(i), fontsize=12, color='blue')
axes[1, 1].set_title('Flipped Anti-Diagonally')
axes[1, 1].axis('off')
# Hiding empty subplot
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
```

```python
# save arrays
X_h, y_h = [], []
X_v, y_v = [], []
X_d, y_d = [], []
X_ad, y_ad = [], []
for i in range(len(images)):
    image, label = images[i], labels[i]
    flipped_horizontal_image, flipped_horizontal_label = flip_horizontally(image, reconstruct_label(label))
    flipped_vertical_image, flipped_vertical_label = flip_vertically(image, reconstruct_label(label))
    flipped_diagonal_image, flipped_diagonal_label = flip_diagonally(image, reconstruct_label(label))
    flipped_anti_diagonal_image, flipped_anti_diagonal_label = flip_anti_diagonally(image, reconstruct_label(label))
    #
    flipped_horizontal_label = sort_label(flipped_horizontal_label)
    flipped_vertical_label = sort_label(flipped_vertical_label)
    flipped_diagonal_label = sort_label(flipped_diagonal_label)
    flipped_anti_diagonal_label = sort_label(flipped_anti_diagonal_label)
    #
    X_h.append(flipped_horizontal_image)
    y_h.append(flipped_horizontal_label)
    
    X_v.append(flipped_vertical_image)
    y_v.append(flipped_vertical_label)
    
    X_d.append(flipped_diagonal_image)
    y_d.append(flipped_diagonal_label) 
    
    X_ad.append(flipped_anti_diagonal_image)
    y_ad.append(flipped_anti_diagonal_label)
```
